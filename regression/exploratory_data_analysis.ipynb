### Feature analysis

```python
# Calculate the correlation of the features with the target
df.corr()['target'].sort_values(ascending=False)
```

```python
# Calculate the correlation matrix of the dataset
correlation_matrix = df.corr()
correlation_threshold = 0.6
correlation_pairs = {}
for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        if abs(correlation_matrix.iloc[i, j]) > correlation_threshold:
            correlation_pairs[(correlation_matrix.columns[i], correlation_matrix.columns[j])] = correlation_matrix.iloc[i, j]
# skip where both columns are the same
correlation_pairs = {pair: value for pair, value in correlation_pairs.items() if pair[0] != pair[1]}

# sort by correlation value
correlation_pairs = dict(sorted(correlation_pairs.items(), key=lambda item: item[1], reverse=True))

correlation_pairs_df = pd.DataFrame(
    [(pair[0], pair[1], value) for pair, value in correlation_pairs.items()],
    columns=["Column 1", "Column 2", "Correlation"]
)

correlation_pairs_df
```

```python
# Plot the pairplot of the dataset
for i in range(0, len(df.columns), 5):
    sns.pairplot(data=df, x_vars=df.columns[i:i + 5], y_vars=['target'])
```

```python
# Quadratic transformation: x^2
df['6_quadratic'] = df['6'] ** 2

# Plot the transformed feature vs target
sns.scatterplot(data=df, x='6_quadratic', y='target')

# Add titles and labels for clarity
plt.title('Quadratic Transformation of 6 vs Target')
plt.xlabel('6 (Quadratic Transformation)')
plt.ylabel('Target')

plt.show()
```

```python
# Calculate correlation between the squared feature 6 and the target
correlation = df[['6_quadratic', 'target']].corr().iloc[0, 1]

# Display the correlation
print(f"Correlation between squared feature '6' and target: {correlation}")
```
